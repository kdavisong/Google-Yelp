Función para realizar extracción de datos de la API de Census Bureau. Estos datos serán tratados en el libro "EDA_D_ext.ipynb", para más información referirse a este notebook.

```python
'''Descarga de datos'''
url_base = "https://api.census.gov/data/"
años = range(2010, 2024)

for año in años:
    url = f"{url_base}{año}/cps/basic/may?get=HRNUMHOU,PTERNH1O,PEMNTVTY,PEFNTVTY,GTCBSASZ,GESTFIPS,GTCSA,PRCHLD,PRTAGE,PESEX,GTCBSA,HEFAMINC,HRHTYPE&for=state:*"
    r = requests.get(url)
    nom_col = r.json()[0]
    # Verifica si la solicitud fue exitosa
    df = pd.DataFrame(columns = nom_col, data = r.json()[1:])
    file_name = f"Current_Population_Survey{año}.json"
    df.to_json(file_name)
    print(f"Datos para el año {año} descargados y guardados en {file_name}")
```
