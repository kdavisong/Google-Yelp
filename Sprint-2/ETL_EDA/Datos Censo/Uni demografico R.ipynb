```python
for nombre_archivo in os.listdir(os.getcwd()):
    if nombre_archivo.endswith('.json'):
        ruta_archivo = os.path.join(nombre_archivo)
        with open(ruta_archivo, 'r') as archivo_json:
            datos = pd.read_json(archivo_json)
        datos = datos[['GTCBSA', 'state', 'HEFAMINC', 'GTCBSASZ', 'GESTFIPS']]
        for idx, fila in datos.iterrows():
            if fila['GTCBSA'] in df['id_area_metropolitana'].values:
                index = df[df['id_area_metropolitana'] == fila['GTCBSA']].index[0]
                df.at[index, 'total_nivel_ingresos'].append(fila['HEFAMINC'])
                df.at[index, 'instancias'] += 1
            else:
                fila_dict = {
                    'id_area_metropolitana': fila['GTCBSA'],
                    'id_estado' : fila['GESTFIPS'],
                    'total_nivel_ingresos': [fila['HEFAMINC']],
                    'nivel_area_metropolitana': fila['GTCBSASZ'],
                    'instancias': 1
                }
                df = df.append(fila_dict, ignore_index=True)


```

```python
def promedio(lst):
    if len(lst) > 0:
        return round(sum(lst) / len(lst))
    else:
        return None
df = df.merge(df_mapa[['id_area_metropolitana', 'condado']], on='id_area_metropolitana', how='left')
df['ingreso promedio'] = df['total_nivel_ingresos'].apply(promedio)
df['estado'] = df['id_estado'].map(mapa_estados)
```
